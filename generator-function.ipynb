```python
def test_fibonacci(n):
    a,b = 0,1
    for i in range(n):
        yield a  #yield is a generator function that throws the data
        a,b = b,a+b
```

```python
test_fibonacci(10)
```

```python
for i in test_fibonacci(10):
    print(i)
```

```python
def test_fibonacci1():
    a,b = 0,1
    while True:
        yield a  #yield is a generator function that throws the data
        a,b = b,a+b
```

```python
fib = test_fibonacci1()
```

```python
type(fib)
```

```python
for i in range(10):
    print(next(fib))
```

```python
def count_test(n):
    count = 1
    while count < n:
        yield count 
        count = count +1
```

```python
c = count_test(5)
```

```python
for i in c:
    print(i)
```
