# Reduce Function

```python
from functools import reduce
```

```python
l = [1,2,3,3,4,5]
```

```python
reduce(lambda x,y : x+y, l)
```

Only two paremeters can be passed in reduce

```python
reduce(lambda x,y,z : x+y+z, l) #Error because z is missing positional argument
```

```python
reduce(lambda x,y : x+y, []) #Error blank list
```

```python
reduce(lambda x,y : x+y, [1]) #Exceptional one element can be passed but not more than two
```

```python
reduce(lambda x,y:x**y,l)
```

```python
reduce(lambda x,y : x if x>y else y,l)
```
