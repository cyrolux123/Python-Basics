```python
n = int(input('Enter your number'))
sum = 0
counter = 1

while counter <= n:
    sum = sum + counter
    counter = counter +1
print('Total sum is',sum)
```

```python
number = int(input('Enter your number'))
factorial = 1
while(number > 0):
    factorial = factorial * number
    number = number -1
factorial
```

```python
num = int(input('Enter the element'))
a,b = 0,1
counter = 0
while counter < number:
    print(a)
    c = a + b
    a = b
    b = c
    counter = counter + 1
```

```python
word = input("Enter the string for reversal")
reverse = ""
length = len(word)
while(length > 0):
    reverse = reverse + word[length - 1]
    length = length -1
print(reverse)
```
